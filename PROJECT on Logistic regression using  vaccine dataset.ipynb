# CLEANING DATA

```python
#removing has_health_insur because of high null values.
vac_data=vaccine_data.drop(["has_health_insur"],axis=1)
vac_data
```

```python
vac_data.describe()
```

```python
#finding total null values
vac_data.isnull().sum()   
```

```python
#storing all the columns in a variable which needs to be set as 0
col_to_zero=["employment","marital_status","housing_status","income_level","qualification","sick_from_seas_vacc","is_seas_vacc_effective","is_seas_risky","sick_from_h1n1_vacc","is_h1n1_risky","is_h1n1_vacc_effective","is_health_worker","cont_child_undr_6_mnths","chronic_medic_condition","dr_recc_seasonal_vacc","dr_recc_h1n1_vacc","avoid_touch_face","reduced_outside_home_cont","avoid_large_gatherings","wash_hands_frequently","bought_face_mask","contact_avoidance","antiviral_medication","h1n1_awareness","h1n1_worry"]
```

```python
#filling 0 to the columns taken
vac_data[col_to_zero]=vac_data[col_to_zero].fillna(0)
```

```python
vac_data.isnull().sum()
```

```python
#filling mean value to one of column
vac_data["no_of_adults"]=vac_data["no_of_adults"].fillna(vac_data.no_of_adults.mean())
```

```python
#filling median value to one of column
vac_data["no_of_children"]=vac_data["no_of_children"].fillna(vac_data.no_of_adults.median())
```

```python
vac_data.isnull().sum()
```

# creating dummy variables

```python
vac_data=pd.get_dummies(vac_data,drop_first=True)
```

```python
vac_data
```

# Now data is fully cleaned,all the null values are fixed and column with object datatypes are fixed with dummy variables so now data is set to proceed.

# ROC curve-(Reciver Operating Characterstic curve)

#
roc is mainly used because we can set a threshold point from where we can diffrentiate 0 and 1

```python
from sklearn.metrics import roc_auc_score
from sklearn.metrics import roc_curve
```

```python
logis_roc_auc=roc_auc_score(y_test,y_pred)
```

```python
fpr,tpr,thresholds=roc_curve(y_test,model1.predict_proba(x_test)[:,1])
```

```python
plt.plot(fpr,tpr,logis_roc_auc)
```

```python
#setting threshold as o.8 curve 
```

```python
roc_t=LogisticRegression(class_weight="balanced")
roc_t.fit(x_train,y_train)
THRESHOLD=0.8
```

```python
import numpy as np
```

```python
y_pred_roc=np.where(roc_t.predict_proba(x_test)[:,1]>THRESHOLD,1,0)
```

```python
accuracy_score(y_test,y_pred_roc)
```
