```python
x_ind=mobile.drop("price_range",axis=1)
y_dep=mobile.price_range
```

```python
from sklearn import model_selection
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_ind, y_dep, test_size=0.3, random_state=2)
```

```python
plt.figure(figsize=(10,5))
sns.stripplot(x='clock_speed',y='ram',data=mobile,palette='spring',hue='price_range');
```

```python
from sklearn.preprocessing import StandardScaler
se=StandardScaler()
```

```python
x_train=se.fit_transform(x_train)
x_test=se.fit_transform(x_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
mobile.shape
```

```python
import numpy as np
```

```python
np.sqrt(2000)
```

```python
KNN=KNeighborsClassifier(n_neighbors=45,p=3,metric="euclidean")
```

```python
KNN.fit(x_train,y_train)
```

```python
y_pred=KNN.predict(x_test)
y_pred
```

```python
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
```

```python
confusion_matrix(y_test,y_pred)
```

```python
accuracy_score(y_test,y_pred)
```

```python
error_rate=[]
for i in range(1,50):
    knn_new=KNeighborsClassifier(n_neighbors=i)
    knn_new.fit(x_train,y_train)
    y_pred1=knn_new.predict(x_test)
    error_rate.append(np.mean(y_pred1 != y_test))
```

```python
error_rate
```

```python
plt.figure(figsize=(15,10))
plt.plot(range(1,50),error_rate,color='blue',linestyle='dashed',
        marker='o',markerfacecolor='red')
```

```python
KNN=KNeighborsClassifier(n_neighbors=47,p=3,metric="euclidean")
```

```python
KNN.fit(x_train,y_train)
```

```python
y_pred=KNN.predict(x_test)
```

```python
accuracy_score(y_test,y_pred)
```
