# Analyzing missing values

```python
#only features with null values
features_with_na=[feature for feature in df if df[feature].isnull().sum()]
for feature in features_with_na:
    print(feature,df[feature].isnull().sum())
```

```python
#checking whether missing values have any relation with sales price or not.
#assingning one for null values and zero for non-null values and making groupby for all the features.
plt.figure(figsize=(18,80))
for i,feature in enumerate(features_with_na,1):
    plt.subplot(19,3,i)
    data = df.copy()
    data[feature] = np.where(data[feature].isnull(), 1, 0)
    data.groupby(feature)["SalePrice"].median().plot.bar()
    plt.ylabel("SalePrice")
plt.show()
```

Here the relation between the missing values and the dependent variable is clearly visible,So We need to replace these nan values with something meaningful.

# FEATURE ENGINEERING

Handling missing values

```python
## Now lets check for numerical variables that contains missing values
```

```python
numerical_nan_values=[feature for feature in numerical_features if df[feature].isnull().sum()>1]
numerical_nan_values
```

```python
#Here only continious features have null values but none of the discrete features have null values.
```

```python
df[numerical_nan_values].isnull().sum()
```

```python
#Replacing the numerical Missing Values
for feature in numerical_nan_values:
    df[feature]=df[feature].fillna(df[feature].median())
```

```python
df[numerical_nan_values].isnull().sum()
```

```python
## Now lets check for catogorical variables that contains missing values
categorical_with_nan=[feature for feature in categorical_features if df[feature].isnull().sum()]
categorical_with_nan
```

```python
#Replacing null values in catogorical values with a new label "Unknown" and finally grouping those unknown data as a indivudual catogory in data encoding. 
```

```python
## Replace missing value with a new label
for feature in categorical_with_nan:
    df[feature]=df[feature].fillna("Unknown")
```

```python
df.head(10)
```

```python
# Treating Temporal Variables (Date Time Variables)
for feature in year_feature:
    if feature!="YrSold":
        df[feature]=df['YrSold']-df[feature]
```

```python
df.head(10)
```

```python
#droping Alley & MiscFeature, poolqc columns which has 1369,1406 & 1453  null values,those are above 90%.
#Also dropping id column because it is a unique feature,which dosn't contribute much in prediction.
df.drop(["Alley","MiscFeature","PoolQC","Id"],axis=1,inplace=True)
```

```python
df
```

```python
#Checking any null values in data
```

```python
for feature in features_with_na:
    if feature!="Alley" and feature!="MiscFeature" and feature!="PoolQC":
        print(feature,df[feature].isnull().sum())
```
